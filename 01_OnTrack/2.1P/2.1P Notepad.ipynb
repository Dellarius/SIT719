```python
# This is a made up senario to demonstrate common Pandas DataFrame Manipulations, and matlplotlib visualisations.
# Our dataset is RSSI and SNR readings on a Wireless Network taken every 5 seconds for 10 minutes.

time_index = pd.date_range(start='2025-11-20 12:21:00', periods=120, freq='5s')
data = {
    'RSSI': np.random.randint(-100, -76, size=120),
    'SNR': np.random.randint(0, 40, size=120)
}   
wireless_survey = pd.DataFrame(data, index=time_index)

time_index = time_index.tz_localize('Australia/Brisbane')

one_min_avg = (
    wireless_survey
        .groupby(wireless_survey.index.floor('1min'))
        .mean()
        .round(2)
)
print(one_min_avg)

wireless_survey.plot()
plt.gca().invert_yaxis()
plt.title('Wireless Network Survey - RSSI and SNR over Time')
```
