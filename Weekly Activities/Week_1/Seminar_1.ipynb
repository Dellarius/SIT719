```python
#Q1 V. Create a function to calculate factorial of a number

def function_factorial(x):
    y = 1
    for i in range(1, x + 1):
        y = y * i
    return y

# To Calculate a factorial of x, input your number in () below 
function_factorial(10)
```

```python
#Q3 III. Import dict into pandas dataframe

exam_data = {
    'name': ['Alice', 'Dee', 'Kathy', 'Homes', 'Boris', 'Hanson', 'Bourne', 
            'Matt', 'Jason', 'Jarvis'], 
    'score': [115, 7, 15.5, np.nan, 9, 19, 15.5, np.nan, 8.5, 19.5], 
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
} 
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

student_data = pd.DataFrame(exam_data, index=labels)
student_data
```

```python
#Q3 III. a) Change Dee to Eve

student_data['name'] = student_data['name'].replace('Dee', 'Eve')
student_data
```

```python
#Q3 III. b) Find rows with missing values

student_data[(student_data.isna().any(axis=1))]
```

```python
#Q3 III. c) Show rows where the number of attempts is less than 2 and score greater than 15

student_data[(student_data['attempts'] < 2) & (student_data['score'] > 15)]
```

```python
#Q3 III. d) Replace NaN with 10

student_data.fillna(value=10)
```
